```python
def valider_une_ligne(line_index):
    indices = lignes[line_index]
    valeurs = [hexagone.valeurs[i] for i in indices]
    if any(v is None for v in valeurs):
        return False
    return sum(valeurs) == 38


def valider_toutes_les_lignes():
    return [valider_une_ligne(i) for i in range(len(lignes))]


def etat_des_lignes():
    resultats = []
    for i, indices in enumerate(lignes):
        valeurs = [hexagone.valeurs[j] for j in indices]
        #print(valeurs)
        somme = None if any(v is None for v in valeurs) else sum(valeurs)
        resultats.append({
            "indices": indices,
            "valeurs": valeurs,
            "somme": somme,
            "correct": (somme == 38)
        })
    return resultats



#print(valider_toutes_les_lignes())
for elem in etat_des_lignes():
    print(elem)

# Mettre à jour les couleurs dans le widget
hexagone.widget.update_line_colours(etat_des_lignes())
hexagone.dessiner_hex()
```
